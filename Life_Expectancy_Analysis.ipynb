# Data Analysis of Life Expectancy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/life-expectancy/WHOSIS_000001WHOSIS_000015WHOSIS_000002WHOSIS_000007(1).csv')
df.head(50)
```

```python
df.duplicated().sum()

print(df.columns)
```

```python
df.describe()
```

# Leading Question
## Are there significant differences in life expectancy between males and females globally and within different income groups?

```python
# dict to hold gender means every year, key is the year and value is the mean
gender_means = {}
for year, group in df.groupby('Year'):
    # creates a dictionary for each year
    gender_means[year] = {
        # 2 key-value pairs (both values are the means)
        'Male': group['Life expectancy at birth (years)  Male'].mean(),
        'Female': group['Life expectancy at birth (years)  Female'].mean()
    }

# list for all the years in ascending order
years = sorted(gender_means.keys())

# accessing the dictionary corresponding to the year
# gather all the male and female means into 2 lists
male_means = [gender_means[year]['Male'] for year in years]
female_means = [gender_means[year]['Female'] for year in years]

# Create an array of years for x-axis, 
x = np.arange(len(years))
width = 0.35  # Width of each bar

# Plot bars for males and females
plt.bar(x - width/2, male_means, width, label='Male', color='cyan')
plt.bar(x + width/2, female_means, width, label='Female', color='orange')

plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy (Male vs. Female) Over the Years')
plt.xticks(x, years)
plt.legend()

plt.tight_layout()
plt.show()
```

# General Observations
1. Overalll females have a longer life expectancy
2. Life expectancy also seems to be increasing every year for both males and females.
3. The exact gap difference is hard to see, another graph would be beneficial to visualize this.

```python
df_sorted = df.sort_values(by='Year')
years = df_sorted['Year'].unique()

male_life_expectancy = df.groupby('Year')['Life expectancy at birth (years)  Male'].mean()
female_life_expectancy = df.groupby('Year')['Life expectancy at birth (years)  Female'].mean()

grouped_by_year = df.groupby('Year')

avg_gap = female_life_expectancy - male_life_expectancy

print(avg_gap)

# width by height inches
plt.figure(figsize=(10, 6))
plt.plot(years, avg_gap, c='red', label='Average Gap')

plt.xlabel('Year')
plt.ylabel('Life Expectancy Gap')
plt.title('Life Expectancy Between Males and Females Over the Years')
plt.grid(True)
plt.legend()

plt.show()
```

# Average Gap Observations
1. The gap increased from 2000 to 2010, and now the gap seems to be decreasing by a little
2. The difference in life expectancy is around 4-5 years.

# Differences In Income Groups

```python
income_groups = ['Low-income', 'Lower-middle-income', 'Upper-middle-income', 'High-income']

male_data = {}
female_data = {}

# loop through income groups
for group in income_groups:
    # get only the rows belonging to the current income group
    group_data = df[df['World Bank income group']==group]
    years = group_data['Year'].tolist()
    # get the life expectancy at birth for the current income group
    male_life_expectancy = group_data['Life expectancy at birth (years)  Male'].tolist()
    female_life_expectancy = group_data['Life expectancy at birth (years)  Female'].tolist()
    
    # combined years, male and female life expectancy into tuples and sorts based on the year
    # for each year we will have a tuple that looks like
    #   - [(2015, 61.0, 65.4), (2019, 63.0, 67.2)]
    
    sorted_data = sorted(zip(years, male_life_expectancy, female_life_expectancy))
    # takes sorted data and unpacks into separate lists
    years, male_life_expectancy, female_life_expectancy = zip(*sorted_data)
    
    # place the life expectancies into the data dicts at corresponding income group
    # male and female life_expectancy are lists
    male_data[group] = male_life_expectancy
    female_data[group] = female_life_expectancy
    
    print(f'income_group: {group} and data (sorted by year): {male_data[group]}')
    
print(years)

plt.figure(figsize=(12, 10))
for group in income_groups:
    plt.plot(years, male_data[group], label=f'{group} (Male)')
    plt.plot(years, female_data[group], label=f'{group} (Female)', linestyle='--')  # Use a different linestyle for females

plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('Life Expectancy (Male vs. Female) by Income Group Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

# Observations
1. Females with high incomes have the highest life expectancy
2. Males with low incomes have lowest life expectancy
3. Over the years females in low income groups have a life expectancy close to males with lower-middle income.
4. From 2016 and onwards females in upper-middle income groups have higher life expectancy than males with high income

```python
income_groups = ['Low-income', 'Lower-middle-income', 'Upper-middle-income', 'High-income']

# Create subplots with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 12))

# Loop through income groups
for i, group in enumerate(income_groups):
    # Get only the rows belonging to the current income group
    group_data = df[df['World Bank income group'] == group]
    
    years = group_data['Year'].tolist()
    male_life_expectancy = group_data['Life expectancy at birth (years)  Male'].tolist()
    female_life_expectancy = group_data['Life expectancy at birth (years)  Female'].tolist()
    
    sorted_data = sorted(zip(years, male_life_expectancy, female_life_expectancy))
    years, male_life_expectancy, female_life_expectancy = zip(*sorted_data)
    
    # determine where to place each subgraph in  the 2x2 graph
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    
    ax.plot(years, male_life_expectancy, label=f'{group} (Male)')
    ax.plot(years, female_life_expectancy, label=f'{group} (Female)', linestyle='--')
    
    ax.set_xlabel('Year')
    ax.set_ylabel('Life Expectancy at Birth (years)')
    ax.set_title(f'Life Expectancy (Male vs. Female) in {group}')
    ax.legend()
    ax.grid(True)

# Adjust subplot layout
plt.tight_layout()

# Show the subplots
plt.show()
```

# Observations
1. The gap between male and female life expectancy increases as the income increases
> lower income has a smaller gap compared to low-middle income.
2. The biggest difference in life expectancy between males and females is in the high income group 


# Comparing Female Data


```python
data = {group: [] for group in income_groups}
years = []

for group in income_groups:
    group_data = df[df['World Bank income group'] == group]
    
    years = group_data['Year'].tolist()
    female_life_expectancy = group_data['Life expectancy at birth (years)  Female'].tolist()
    
    sorted_data = sorted(zip(years, female_life_expectancy))
    years, female_life_expectancy = zip(*sorted_data)
    data[group] = female_life_expectancy
    
x = np.arange(len(years))

bar_width = 0.2

fig, ax = plt.subplots(figsize=(12, 8))

for i, group in enumerate(income_groups):
    ax.bar(x+i * bar_width, data[group], width=bar_width, label = f'{group} (Female)')
   
ax.set_xlabel('Year')
ax.set_ylabel('Life Expectancy At Birth (years)')
ax.set_title('Female Life Expectancy by Income Group Over the Years')
ax.set_xticks(x + (bar_width * (len(income_groups) - 1)) / 2)
ax.set_xticklabels(years)
ax.legend()
ax.grid(True)

# Show the grouped bar graph
plt.show()
    
```

# Observations
1. Aside from low income group females, there doesn't seem to be a big growth in life expectancy over the years in the other groups
2. Life expectancy for females in high income groups looks constant

# Comparing Male Data

```python
data = {group: [] for group in income_groups}
years = []

for group in income_groups:
    group_data = df[df['World Bank income group'] == group]
    
    years = group_data['Year'].tolist()
    male_life_expectancy = group_data['Life expectancy at birth (years)  Male'].tolist()
    
    sorted_data = sorted(zip(years, male_life_expectancy))
    years, male_life_expectancy = zip(*sorted_data)
    data[group] = male_life_expectancy
    
x = np.arange(len(years))

bar_width = 0.2

fig, ax = plt.subplots(figsize=(12, 8))

for i, group in enumerate(income_groups):
    ax.bar(x+i * bar_width, data[group], width=bar_width, label = f'{group} (Male)')
   
ax.set_xlabel('Year')
ax.set_ylabel('Life Expectancy At Birth (years)')
ax.set_title('Male Life Expectancy by Income Group Over the Years')
ax.set_xticks(x + (bar_width * (len(income_groups) - 1)) / 2)
ax.set_xticklabels(years)
ax.legend()
ax.grid(True)

# Show the grouped bar graph
plt.show()
    
```

# Summary Analysis

While exploring this dataset, several notable patterns and trends emerged. 

1. **Income Group and Life Expectancy**
    * There exists a correlation between income groups and life expectancy
    * Individuals in higher income groups exibit a higher life expectancy compared to those in lower income groups
    * The highest life expectancy is observed in the high-income group, while the lowest is found in the low-income group.
    
2. **Gender Disparities**
    * Females consistently outlive males by an average of 4-5 years each year from 2000 to 2019.
    * This gender gap is greatest in the high-income group and smallest in the low-income group.
    
3. **Trends**
    * Over time the gender gap in life expectancy has shown a decreasing trend, particularly after 2010. This suggests that the gap is narrowing even though females continue to have a longer life expectancy.
    
# Inferences

Based on the observed trends, there are several inferences that can me made.

1. **Continued Gender Gap Reduction**
    * It seems that the gender gap in life expectancy will continue to decrease as time goes on.
   
2. **Gender Disparity**
    * Despite improvements, females are expected to maintain a higher life expectancy than males in all income groups

3. **Low-Income Groups**
    * In the near future, it is likely than low-income females will surpass the life expectancy of malesin the low-middle income group.
